```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Household Power Consumption datasets

#importing results:
errorValuesPCA = pd.read_csv(
    "./Results/household_power_consumption/Details PCA.csv",
    )
errorValuesPCA.index += 1
errorValuesKPCA = pd.read_csv(
    "./Results/household_power_consumption/Details K-PCA.csv",
    )
errorValuesKPCA.index += 1
errorValuesSVD = pd.read_csv(
    "./Results/household_power_consumption/Details SVD.csv",
    )
errorValuesSVD.index += 1

plt.figure(figsize=(10, 10))
plt.subplot(3,2,1)
plt.plot(errorValuesPCA[['Root Mean Squared Values']], marker='o', linestyle='--', label="PCA")
plt.plot(errorValuesSVD[['Root Mean Squared Values']], marker='o', linestyle='--', label="SVD")
plt.plot(errorValuesKPCA[['Root Mean Squared Values']], marker='o', linestyle='--', label="K-PCA")
plt.legend()
plt.title('Root Mean Squared Error (RMSE)')
plt.xlabel('Number of Dimensions')
plt.grid()

plt.subplot(3,2,2)
plt.plot(errorValuesPCA[['Mean Absolute Values']], marker='o', linestyle='--', label="PCA")
plt.plot(errorValuesSVD[['Mean Absolute Values']], marker='o', linestyle='--', label="SVD")
plt.plot(errorValuesKPCA[['Mean Absolute Values']], marker='o', linestyle='--', label="K-PCA")
plt.legend()
plt.title('Mean Absolute Error (MAE)')
plt.xlabel('Number of Dimensions')
plt.grid()

plt.subplot(3,2,3)
plt.plot(errorValuesPCA[['R Squared Error Values']], marker='o', linestyle='--', label="PCA")
plt.plot(errorValuesSVD[['R Squared Error Values']], marker='o', linestyle='--', label="SVD")
plt.plot(errorValuesKPCA[['R Squared Error Values']], marker='o', linestyle='--', label="K-PCA")
plt.legend()
plt.title('R Squared')
plt.xlabel('Number of Dimensions')
plt.grid()

plt.subplot(3,2,4)
plt.plot(errorValuesPCA[['Compression Ratio']], marker='o', linestyle='--', label="PCA")
plt.plot(errorValuesKPCA[['Compression Ratio']], marker='o', linestyle='--', label="SVD")
plt.plot(errorValuesSVD[['Compression Ratio']], marker='o', linestyle='--', label="K-PCA")
plt.legend()
plt.title('Compression Ratio')
plt.xlabel('Number of Dimensions')
plt.grid()

plt.subplot(3,1,3)
plt.plot(errorValuesPCA[['Variance']], marker='o', linestyle='--')
plt.title('Variance')
plt.xlabel('Number of Dimensions')
plt.grid()
plt.tight_layout()
plt.show()
```
