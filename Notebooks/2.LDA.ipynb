## LDA (Linear Disriminant Analysis):
LDA is a supervised dimensionality reduction technique primarily used for classification tasks. Unlike PCA, which focuses on preserving variance, LDA aims to maximize the separation between multiple classes by finding a projection that best discriminates among classes.
This is a Linear technique
